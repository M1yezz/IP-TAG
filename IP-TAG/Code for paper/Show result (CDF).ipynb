```python
# Show result (CDF)

import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import pandas as pd
import numpy as np
import seaborn as sns

# Read the file
data1 = pd.read_csv("./01architecture_3AE_200.csv",header=None)
#data2 = pd.read_csv("./01architecture_4AE_80.csv",header=None)
#data3 = pd.read_csv("./01architecture_3DNN_700.csv",header=None)
#data4 = pd.read_csv("./01architecture_4DNN_400.csv",header=None)

x1 = data1.iloc[:,3]
#x2 = data2.iloc[:,3]
#x3 = data3.iloc[:,3]
#x4 = data4.iloc[:,3]

sns.set_theme(style="whitegrid") 

# Set the canvas size
fig = plt.gcf()
fig.set_size_inches(8,6)

# Set the font size
plt.xticks(fontsize=13,fontweight='bold') 
plt.yticks(fontsize=13,fontweight='bold')

# Set the thickness of the coordinate axes
ax=plt.gca(); # Get the handle of the axis
ax.spines['bottom'].set_linewidth(0.6); # Bottom Axis
ax.spines['left'].set_linewidth(0.6); # Left Axis
ax.spines['right'].set_linewidth(0.6); # Right Axis
ax.spines['top'].set_linewidth(0.6) # Top Axis
# Set the color of the coordinate axes
ax.spines['bottom'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['right'].set_color('black')
ax.spines['top'].set_color('black')

# Set the scale spacing
y_major_locator=MultipleLocator(0.1)
ax.yaxis.set_major_locator(y_major_locator)

# Set the range of the coordinate axes
#plt.xlim(0, 3)
plt.ylim(0, 1)
#sns.set_context("paper")

#sns.distplot(x2,color='000',label='multi-task AE with 7 layers',hist=False,hist_kws={'cumulative': True, 'density': True}, kde_kws={'cumulative': True})
#sns.distplot(x4,color='000',label='DNN with 4 layers',hist=False,hist_kws={'cumulative': True, 'density': True}, kde_kws={'cumulative': True,'linestyle':'--'})
sns.distplot(x1,color='000',label='multi-task AE with 5 layers',hist=False,hist_kws={'cumulative': True, 'density': True}, kde_kws={'cumulative': True,'linestyle':'-.'})
#sns.distplot(x3,color='000',label='DNN with 3 layers',hist=False,hist_kws={'cumulative': True, 'density': True}, kde_kws={'cumulative': True,'linestyle':':'})

font1 = {'family' : 'Times New Roman','weight' : 'bold','size': 12}
plt.legend(prop=font1)

# Set the description
plt.xlabel("Positioning Error (m)",fontsize=14,fontweight='bold',labelpad=6)
plt.ylabel('Cumulative Distribution Function (CDF)',fontweight='bold',fontsize=14,labelpad=6)
#plt.grid(ls='--',c='grey')
plt.savefig("./01_architecture.pdf")
plt.show()
```
